Нужно научиться определять рыночную стоимость объектов недвижимости. Наша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

**Промежуточные цели исследования:**

- Определить факторы, которые больше всего влияют на стоимость квартиры. Изучить, зависит ли цена от площади, числа комнат, удалённости от центра. - Выбрать 10 населённых пунктов с наибольшим числом объявлений. Рассчитать среднюю цену квадратного метра в этих населённых пунктах. Выделить населённые пункты с самой высокой и низкой стоимостью жилья.
- Определить область на карте, которую можем считать центром.

Выделить факторы, которые влияют на стоимость квартиры (число комнат, этаж, удалённость от центра, дата размещения объявления). Сделать вывод, отличаются ли они от общих выводов по всему городу.

Создадим функцию для расчёта межквартильного размаха и усов, она нам не раз пригодится.

```python
# Функция для расчёта межквартильного размаха и усов
def calculate_iqr_whiskers(column):
    column_min = data[column].min()
    column_iqr = data[column].quantile(0.75) - data[column].quantile(0.25)
    left_whisker = max(column_min, data[column].quantile(0.25) - 1.5*column_iqr)
    right_whisker = data[column].quantile(0.75) + 1.5*column_iqr
    
    return column_iqr, left_whisker, right_whisker
```

```python
for column in ['total_area', 'last_price', 'rooms', 'ceiling_height']:
    parameters = calculate_iqr_whiskers(column)
    print(f'Размах усов для {column}: от {parameters[1]:n} до {parameters[2]:n}')
```

```python
for column in ['total_area', 'last_price', 'rooms', 'ceiling_height']:
    parameters = calculate_iqr_whiskers(column)
    data.hist(column, bins=20, range=(parameters[1], parameters[2]))
```

Обновим формулу для расчёта межквартильного размаха усов:

```python
# Функция для расчёта межквартильного размаха и усов
def calculate_iqr_whiskers_spb(column):
    column_min = cleaned_data_center[column].min()
    column_iqr = cleaned_data_center[column].quantile(0.75) - cleaned_data_center[column].quantile(0.25)
    left_whisker = max(column_min, cleaned_data_center[column].quantile(0.25) - 1.5*column_iqr)
    right_whisker = cleaned_data_center[column].quantile(0.75) + 1.5*column_iqr
    
    return column_iqr, left_whisker, right_whisker
```

```python
for column in ['total_area', 'last_price', 'rooms', 'ceiling_height']:
    parameters = calculate_iqr_whiskers_spb(column)
    print(f'Размах усов для {column}: от {parameters[1]:n} до {parameters[2]:n}')
```
